```python
import sklearn
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
import time
from sklearn.svm import SVC
from sklearn.gaussian_process.kernels import RBF
from sklearn.naive_bayes import GaussianNB
# from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.tree import DecisionTreeClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from pathlib import Path
import shap
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(Path("../data/BIKED_processed.csv"), index_col=0)
reduced=pd.read_csv(Path("../data/BIKED_reduced.csv"), index_col=0)
reduced.BIKESTYLE=pd.Categorical(reduced.BIKESTYLE)
reduced['BIKESTYLECAT']=reduced.BIKESTYLE.cat.codes
classes=reduced["BIKESTYLECAT"]
inputvals=df
target=reduced.BIKESTYLE
```

```python
#Train model based on test set and evaluate performance on the test data, return score
def evaluatescore(numtests, data, method,train_size):
    st=time.time()
    score=0
    inputvals=data[0]
    targetvals=data[1]
    for i in range(numtests):
        X_train, X_test, y_train, y_test = train_test_split(inputvals, targetvals, train_size=train_size, shuffle=True)
#         X_train, X_test, y_train, y_test = customtts(inputvals, targetvals, train_size=train_size) #This line for custom train-test split
        if method=="svc":
            model = SVC(kernel="linear", C=0.025)
        if method=="nb":
            model=GaussianNB()
        if method=="Kn":
            model=KNeighborsClassifier(5)
        if method=="dt5":
            model=DecisionTreeClassifier(max_depth=5)
        if method=="dt8":
            model=DecisionTreeClassifier(max_depth=8)
        if method=="rf":
            model=RandomForestClassifier(max_depth=7, n_estimators=50)
        if method=="ada":
            model=AdaBoostClassifier()
        if method=="gpc":
            model=GaussianProcessClassifier(1.0 * RBF(1.0))
#         if method=="qda":
#             model=QuadraticDiscriminantAnalysis()
        if method=="mlp1":
            model=MLPClassifier(hidden_layer_sizes=(200), max_iter=5000)
        if method=="mlp2":
            model=MLPClassifier(hidden_layer_sizes=(200, 200), max_iter=5000)
        if method=="mlp3":
            model=MLPClassifier(hidden_layer_sizes=(200, 200, 200), max_iter=5000)
        if method=="mlp4":
            model=MLPClassifier(hidden_layer_sizes=(200, 200, 200), max_iter=5000)
        model.fit(X_train, y_train)
        score=score+100*model.score(X_test, y_test)/numtests
    print(method+" evaluation time: " + str((time.time() - st)))
    return score
#Test all methods to generate method comparison dataframe
def testmethods(data,train_size):
    arow=pd.DataFrame({"K neighbors": evaluatescore(1,data,"Kn",train_size)},index=[len(data[0])])
    arow=arow.join(pd.DataFrame({"Support Vector Classification": evaluatescore(10,data,"svc",train_size)},index=[len(data[0])]))
    arow=arow.join(pd.DataFrame({"Naive Bayes": evaluatescore(10,data,"nb",train_size)},index=[len(data[0])]))
    arow=arow.join(pd.DataFrame({"Depth 5 Decision Tree": evaluatescore(100,data,"dt5",train_size)},index=[len(data[0])]))
    arow=arow.join(pd.DataFrame({"Depth 8 Decision Tree": evaluatescore(100,data,"dt8",train_size)},index=[len(data[0])]))
    arow=arow.join(pd.DataFrame({"Random Forest": evaluatescore(100,data,"rf",train_size)},index=[len(data[0])]))
    arow=arow.join(pd.DataFrame({"AdaBoost": evaluatescore(10,data,"ada",train_size)},index=[len(data[0])]))
    arow=arow.join(pd.DataFrame({"Gaussian Process Classifier": evaluatescore(1,data,"gpc",train_size)},index=[len(data[0])]))
#    arow=arow.join(pd.DataFrame({"Quadratic Discriminant Analysis": evaluatescore(10,data,"qda",train_size)},index=[len(data[0])]))
    arow=arow.join(pd.DataFrame({"3-layer Neural Net": evaluatescore(5,data,"mlp1",train_size)},index=[len(data[0])]))
    arow=arow.join(pd.DataFrame({"4-layer Neural Net": evaluatescore(5,data,"mlp2",train_size)},index=[len(data[0])]))
    arow=arow.join(pd.DataFrame({"5-layer Neural Net": evaluatescore(5,data,"mlp3",train_size)},index=[len(data[0])]))
    arow=arow.join(pd.DataFrame({"6-layer Neural Net": evaluatescore(5,data,"mlp4",train_size)},index=[len(data[0])]))

    return arow
```

```python
#Evaluate scores for several test sizes for several methods

accuracydf=pd.DataFrame()
testvals=[200,600,1200,2000,3000]
for i in testvals:
    arow=testmethods([inputvals,target],i)
    accuracydf=pd.concat([accuracydf,arow])
accuracydf.index=testvals
print(accuracydf)
ax=accuracydf.plot.line(ylim=(50,100))
ax.set_xlabel("Number of Training Samples")
ax.set_ylabel("Classification Accuracy (%)")
ax.legend(bbox_to_anchor=(1.02,1.42),ncol=2, fancybox=True, shadow=False, frameon=False)
```
